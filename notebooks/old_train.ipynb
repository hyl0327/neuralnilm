```python
def load_activations_and_mains(filename, appliances, windows, sample_period):
    """
    Parameters:
        - windows[building] (dict)
    Returns:
        - activations['appliance'][building] (dict)
        - mains[building] (dict)
    -------
    
    """
    data = DataSet(filename)
    
    activations = {}
    for appliance in appliances:
        activations[appliance] = {}
    mains = {}
    
    for building, window in windows.iteritems():
        data.set_window(start=window[0], end=window[1])
        
        print('Loading building {} ...'.format(building))
        elec = data.buildings[building].elec
        mains_elec = elec.mains()
        
        # mains
        print('\t- Loading mains ... ', end='')
        mains_elec_good_sections = mains_elec.good_sections()
        if len(mains_elec_good_sections) != 0:
            mains[building] = mains_elec.power_series_all_data(sample_period=sample_period,
                                                               sections=mains_elec_good_sections)
            print('got {} samples'.format(mains[building].shape[0]))
        else:
            print('none, skipping this building ...')
            continue
        
        # activations
        for appliance in appliances:
            print('\t- Loading activations for {} ... '.format(appliance), end='')
            try:
                activations[appliance][building] = elec[appliance].get_activations(sample_period=sample_period)
                print('got {}'.format(len(activations[appliance][building])))
            except KeyError:
                print('none'.format(appliance))
    
    return (activations, mains)
```

```python
activations, mains = load_activations_and_mains(DATA_FILENAME, APPLIANCES, WINDOWS, SAMPLE_PERIOD)
```

```python
for main in mains.itervalues():
    main.plot()
    plt.show()
```
